```python
# increase transfers

import os
import numpy as np
import pandas as pd
import csv
resultfolddir="/mnt/louisayu/nfs_share2/embryo/code/final/fropredict/result/"
resultdir= [j for j in os.listdir(resultfolddir) if not '.' in j]
pati_list=[j for j in os.listdir(resultfolddir+resultdir[0]) if '.xlsx' in j]

for method in resultdir:
    clinical=[]

    for k in pati_list:  #calculate which rank is the true live birth one
        p=pd.read_csv(resultfolddir+method+'/'+k)
        if p.shape[0]>1 and 1 in list(p['true']):
            clinical.append(str(p[p['true']==1].index.to_list()).replace('[','').replace(']',''))   

    embryo2live=[]
    for k in pati_list:  #calculate the live birth index
        p=pd.read_csv(resultfolddir+method+'/'+k)
        if p.shape[0]>1 and 1 in list(p['true']):
            for j in range(p.shape[0]):
                if p['true'][np.argsort(-p['pred'].to_numpy(), axis=0)[j]]==1:
                    embryo2live.append(j)
                    break         



    print(method,clinical,embryo2live)

    with open("/mnt/louisayu/nfs_share2/embryo/code/final/fropredict/result/"+method+"_transfer_result.csv", "w", newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['clinical','embryo2live'])
        for j in range(len(clinical)):
            content = [clinical[j], embryo2live[j]]
            writer.writerow(content)   
```
